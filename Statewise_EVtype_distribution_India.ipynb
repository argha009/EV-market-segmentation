```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### EV Wheeler type vs chargers dataset

```python
df = pd.read_csv('EV Stats-1.csv',skipfooter=5)
df
```

```python
plt.figure(figsize=(8,8))

sns.barplot(data=df,y=df.State.sort_values(ascending=True),x='Two Wheelers (Category L1 & L2 as per Central Motor Vehicles Rules')

plt.xlabel('Two Wheelers (Category L1 & L2 as per Central Motor Vehicles Rules)')
plt.ylabel('States')
plt.title('Statewise 2W (L1 & L2 category) EV distribution in India')

plt.savefig('1.png')
```

```python
plt.figure(figsize=(8,8))

sns.barplot(data=df,y=df.State.sort_values(ascending=True),x='Two Wheelers (Category L2 (CMVR))')

plt.xlabel('Two Wheelers (Category L2 (CMVR))')
plt.ylabel('States')
plt.title('Statewise 2W (L2 category-CMVR) EV distribution in India')

plt.savefig('2.png')
```

```python
plt.figure(figsize=(6,6))

sns.barplot(data=df,y=df.State.sort_values(ascending=True),x='Two Wheelers (Max power not exceeding 250 Watts)')

plt.xlabel('Two Wheelers (Max power not exceeding 250 Watts)')
plt.ylabel('States')
plt.title('Statewise 2W (max power < 250 Watts)EV distribution in India')
plt.savefig('3.png')
```
